```python
#----------------Visualisasi Gambar1
#Menjumlahkan total perhitungan berdasarkan nilai platform
#Membuat data frame baru

g1 = df.groupby( "Platform" ).count()
g1.drop(columns=['Name', 'Year', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales'], inplace=True)
g1.reset_index(drop=True, inplace=True)
g1.rename({'Rank' : 'Total'}, axis=1)
g1 = pd.DataFrame(g1)
```

```python
#Membuat frame untuk jenis dari platform dan mencari nilai indikaasi persamaan pada platform 
#Membuat data frame baru

df1 = pd.DataFrame([[k,v.values]
                        for k,v in df.groupby('Platform').groups.items()], 
                      columns=['Platform','Jumlah'])
df1 = pd.DataFrame(df1)
```

```python
#Membuat data frame baru

df_data = pd.concat([df1, g1], axis=1)
df_data
```

```python
#Visualisasi data.

plt.figure(figsize = (9,6))
plt.bar(x = df_data['Platform'], height=df_data['Rank'])
plt.title("Platform presented in the dataset")
plt.xlabel("Platform")
plt.ylabel("Game Released")
plt.xticks(rotation = 90)
```

```python
#Melakukan aggregate data - NA Sales

Na_Sales = df.groupby(['Platform']).agg({'NA_Sales': ['sum']})
Na_Sales.columns = ['Na_Sales_Sum']
Na_Sales = Na_Sales.reset_index()
df_NaSales = pd.DataFrame(Na_Sales)
```

```python
#Melakukan aggregate data - EU Sales

Eu_Sales = df.groupby(['Platform']).agg({'EU_Sales': ['sum']})
Eu_Sales.columns = ['Eu_Sales_Sum']
Eu_Sales = Eu_Sales.reset_index()
df_EuSales = pd.DataFrame(Eu_Sales)
```

```python
#Melakukan aggregate data - JP_Sales

Jp_Sales = df.groupby(['Platform']).agg({'JP_Sales': ['sum']})
Jp_Sales.columns = ['JP_Sales_Sum']
Jp_Sales = Jp_Sales.reset_index()
df_JpSales = pd.DataFrame(Jp_Sales)
```

```python
#Melakukan aggregate data - Other_Sales

Other_Sales = df.groupby(['Platform']).agg({'Other_Sales': ['sum']})
Other_Sales.columns = ['Other_Sales_Sum']
Other_Sales = Other_Sales.reset_index()
Other_Sales = pd.DataFrame(Other_Sales)
```

```python
#Menggabungkan seluruh dataframe yang telah dilakukan aggregate

df_data_agg = Na_Sales.merge(Eu_Sales,on='Platform').merge(Jp_Sales,on='Platform').merge(Other_Sales, on='Platform')
df_data_agg
```

```python
#Membuat visualisasi group bar. 

fig, ax = plt.subplots()

platform = df_data_agg['Platform']
x = np.arange(len(platform))

width = 0.15

# Membuat bar chart
ax.bar(x - 3*width/2, df_data_agg['Na_Sales_Sum'], width, label='Na_Sales', color='#E8ED91')
ax.bar(x - width/2, df_data_agg['Eu_Sales_Sum'], width, label='Eu_Sales', color='#929467')
ax.bar(x + width/2, df_data_agg['JP_Sales_Sum'], width, label='Jp_Sales', color='#907245')
ax.bar(x + 3*width/2, df_data_agg['Other_Sales_Sum'], width, label='Others_Sales', color='#00C9B0')

ax.set_ylabel('Sales')
ax.set_title('Sales per platform')

ax.set_xticks(x) 
ax.set_xticklabels(platform.astype(str).values, rotation=45)

ax.legend()

plt.show()
```

```python
#---------------------Visualisasi Gambar 2
#Visualisasi data.

plt.figure(figsize = (9,6))
plt.bar(x = df_data['Genre'], height=df_data['Rank'])
plt.title("Genre Presented in the dataset")
plt.xlabel("Genre")
plt.ylabel("Game Released")
plt.xticks(rotation = 90)
```
